```python
import matplotlib.pyplot as plt
import numpy as np
```

###### Some data

Create some data

`np.arrange()` will create a series

```python
x = 0.5 + np.arange(8)     # create 0.5-7
y= [1,3,5,8,3,6,4,6]       # random
y2= [3,5.6,9.4,9.1,4.6,7.7,8.1,9]       # random (examples later need three plots)
```

### Making a plot
using 

`fig, ax = plt.subplots()`

and

`plt.show()`

and make your plot
eg bar 

```python
fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot(x,y2)
#ax.scatter(x,y)
#ax.fill_between(x,y,y2)
#ax.stackplot(x,y,y2)
#ax.stairs(y)
#ax.stem(x,y)



plt.show()
```

#### set()
Setting parametes on plot using 

`set()`

it can take parameters `xlim`, `xlabel`, `title` and many more

```python
fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot(x,y2)

ax.set(xlabel="x axis label", ylabel="y", title="hey mom")



plt.show()
```

#### The axis functions
Or we can use the functions in axis

`set_title`
`set_xlabel`
`set_ylabel`

```python
fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot(x,y2)

ax.set_title("yo mamma")
ax.set_xlabel("x here")
ax.set_ylabel("y")



plt.show()
```

#### Format the labels
Use a formatter function to format the labels
using 

`ax.xaxis.set_major_formatter(formatter_fun)`
###### The function that will do the formatting
It takes two parameters the value on the axis and the position of the tick
It should return text as it should appear

```python
def currency_formatter(x,pos):      # value on x axis and position
    s = f"${x}M"                    # string will return what ever x is with a $ in front and what M for Millon after it.
    return s
```

###### Pass the function to 
`ax.xaxis.set_major_formatter()`

```python
fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot(x,y2)

ax.set_title("yo mamma")
ax.set_xlabel("x here")
ax.set_ylabel("y")

ax.xaxis.set_major_formatter(currency_formatter)

labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')            # put labels at an angle 45 degrees


plt.show()
```

### Add text
Using text()

`ax.text(x, y, "text to add", fontsize=10, verticalalignment="center")`

```python
fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot(x,y2)

ax.set_title("yo mamma")
ax.set_xlabel("x here")
ax.set_ylabel("y")

ax.text(1, 8, "blah", fontsize=10, verticalalignment="center")     # text to a position but use annotation instead - next

ax.xaxis.set_major_formatter(currency_formatter)

labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')


plt.show()
```

#### Annotations 
I love these

`ax.annotate('text', (x_pointer, y_pointer), xytext=(x, y)) `
            

```python
fig, ax = plt.subplots()

ax.bar(x,y)
ax.plot(x,y2)

ax.set_title("yo mamma")
ax.set_xlabel("x here")
ax.set_ylabel("y")

#ax.text(1, 8, "blah", fontsize=10, verticalalignment="center")
ax.annotate('max', (3.2, 8), xytext=(1, 9),arrowprops=dict(facecolor='black', shrink=0.05))        # add annotation instead of text

ax.xaxis.set_major_formatter(currency_formatter)

labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right')


plt.show()
```
