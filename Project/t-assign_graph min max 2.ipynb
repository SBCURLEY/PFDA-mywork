### Plot the data
***

Line Graph 
Purpose: To show how the Dry Bulb Temperature Celsius changes over 24 hours Lecture Notes and Reference 1
- x-axis = Time  "reportEndDateTime"
- y-axis = Dry Bulb Temperature Celsius  "dryBulbTemperature_Celsius"


```python
df["reportEndDateTime"] = pd.to_datetime(df["reportEndDateTime"], errors="coerce")      # convert reportEndDateTime to datetime & specify errors with NaT Reference 2 & 3
# print (df.dtypes)                                                                     # ensuring date & time is set for easier formatting of x-axis

x = df["reportEndDateTime"]
y = df["dryBulbTemperature_Celsius"]

fig, ax = plt.subplots()

ax.plot(x,y)

ax.set_title("Dry Bulb Temperature", fontsize=14,loc="center",
            horizontalalignment="center",verticalalignment="bottom",fontweight="bold",family="monospace")

ax.set_xlabel("End Date & Time",fontsize=10,style="italic",family="monospace")                                         

ax.set_ylabel("Temperature Â°C",fontsize=10,style="italic",family="monospace")  

date_format = DateFormatter("%b %d %H:%M")                        # Displays Month (as October) Date Hours and Minutes Reference 4 and 5
ax.xaxis.set_major_formatter(date_format)

locator = AutoDateLocator()                                       # Automatically adjusts the tick spacing on the x-axis based on the time intervals Reference 6
ax.xaxis.set_major_locator(locator)

ax.grid(axis="y",alpha=0.30)

labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment="right") 

# Adapt from Reference 7 & 8 & 9: 
max_temp = df["dryBulbTemperature_Celsius"].max()                  # Find the max temperature in the dataset column dryBulbTemperature_Celsius
max_temp_index = df["dryBulbTemperature_Celsius"].idxmax()         # find the row number of the max temperature
max_time = df["reportEndDateTime"].iloc[max_temp_index]            # use iloc to find the row from the index from the reportEndDateTime column

ax.annotate("Max", 
            (max_time, max_temp),                                  # Point to annotate
            xytext=(max_time, max_temp - 2.7),                     # Position for the text
            arrowprops=dict(facecolor="black", arrowstyle="-|>"),  # Arrow properties - update arrow style Reference 10
            fontsize=10,
            color="blue") 

min_temp = df["dryBulbTemperature_Celsius"].min()
min_temp_index = df["dryBulbTemperature_Celsius"].idxmin()
min_time = df["reportEndDateTime"].iloc[min_temp_index]

ax.annotate("Min", 
            (min_time, min_temp), 
            xytext=(min_time, min_temp + 2.7),  
            arrowprops=dict(arrowstyle="-|>",facecolor="black"),  
            fontsize=10,
            color="blue") 

# plt.xticks(df['reportEndDateTime'][::5], rotation=45)             # used AutoDateLocator instead

ax.set_ylim(8.5,17)                      

plt.tight_layout()
plt.show()
```
