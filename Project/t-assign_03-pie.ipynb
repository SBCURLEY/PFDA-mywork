### 2. Load the Data
***
Load the dataset from the url

```python
df=pd.read_csv("https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download")
```

### 3. Explore the Data Before Plotting

```python
df.isnull().sum()
```

### 4. Prepare the data
***

```python
df["domain"] = df["Email"].str.split("@").str[1]
df.head(5)
```

Get the number of occurances of each domain

```python
domain_counts = df["domain"].value_counts()
```

### 5. Plot the data
***




```python
nice_colors = ["#ffcc99", "#66b3ff","#99ff99"]
explode = (0.05, 0.05, 0.05)

fig, ax = plt.subplots(figsize=(6, 6))

ax.pie(domain_counts, labels=domain_counts.index, autopct="%1.1f%%", startangle=90, colors=nice_colors, explode=explode, shadow=True)

ax.set_title("Distribution of Domain Names",fontsize=18,loc="center",
            horizontalalignment="center",verticalalignment="bottom",fontweight="bold",family="monospace")

ax.legend(title="Domains",loc="center left",bbox_to_anchor=(1, 0, 0.5, 1))      # legend needs an anchor to not overlap the pie

plt.show()
```
