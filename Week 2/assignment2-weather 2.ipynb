### Plot the data
***

Line Graph 
Purpose: To show how the Dry Bulb Temperature Celsius changes over 24 hours
- x-axis = Time  "reportEndDateTime"
- y-axis = Dry Bulb Temperature Celsius  "dryBulbTemperature_Celsius"


```python
df['reportEndDateTime'] = pd.to_datetime(df['reportEndDateTime'], errors='coerce')      # convert reportEndDateTime to datetime & specify errors with NaT
                                                                                        # https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.to_datetime.html
# print (df.dtypes)                                                                     # ensuring date & time is set for easier formatting of x-axis

x = df["reportEndDateTime"]
y = df["dryBulbTemperature_Celsius"]

fig, ax = plt.subplots()

ax.plot(x,y)

ax.set_title("Dry Bulb Temperature", fontsize=14,loc="center",
            horizontalalignment="center",verticalalignment="bottom",fontweight="bold",family="monospace")

ax.set_xlabel("End Date & Time",fontsize=10,style="italic",family="monospace")                                         

ax.set_ylabel("Temperature (Celsius)",fontsize=10,style="italic",family="monospace")  

date_format = DateFormatter("%B %d %H:%M")      # Displays Month (as October) Date Hours and Minutes
ax.xaxis.set_major_formatter(date_format)       # https://matplotlib.org/stable/gallery/ticks/date_formatters_locators.html

locator = AutoDateLocator()                     # Automatically adjusts the tick spacing on the x-axis based on the time intervals
ax.xaxis.set_major_locator(locator)

ax.grid(axis="y",alpha=0.30)

labels = ax.get_xticklabels()
plt.setp(labels, rotation=45, horizontalalignment='right') 

#plt.xticks(df['reportEndDateTime'][::5], rotation=45)  # used AutoDateLocator instead

ax.set_ylim(8.5,17)                                       

plt.tight_layout()
plt.show()
```
