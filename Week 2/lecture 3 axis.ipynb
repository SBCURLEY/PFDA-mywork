# Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
```

We can create subplots by passing parameters into the subplot function.
it will return a fugure and a 2d array of axes

parameters 
`ncols`, `nrows`

you can set the size using `figsize`

```python
num_cols = 3
num_rows = 2
fig, axs = plt.subplots(ncols=num_cols, nrows=num_rows, figsize=(8,4))      # 2 dimensional array
# axs[0,1].plot([1,2,3],[3,2,1])        # draws a straight line on second box top left to bottom right.
for row in range(num_rows):
    for col in range(num_cols):
        axs[row,col].annotate(f"plot[{row},{col}]", (0.5,0.5), ha="center") # first thing for annotation is a string, position truple horizontal alignment

plt.show()
```

### Example

```python
data = [
    [2,4,5,6,3,5,6,7],
    [9,8,7,6,5,4,3,2],
    [1,2,3,4,5,6,7,8],
    [5,5,5,5,5,5,5,5]
]
x = 0.5 + np.arange(8)
```

```python
num_cols = 2
num_rows = 2
fig, axs = plt.subplots(ncols=num_cols, nrows=num_rows, figsize=(8,4))          # This creates a figure and an array of axes (axs) laid out in a grid with 2 columns and 2 rows 
                                                                                #(num_cols = 2, num_rows = 2).
count = 0                                   # This variable keeps track of which data series is being plotted. It ensures that each subplot gets a different data series.
for row in range(num_rows):     # These nested loops iterate over each position in the grid of subplots (i.e., each combination of row and col).
    for col in range(num_cols):      # For each subplot at position (row, col), it plots the x values (assumed to be defined earlier) against data[count], 
                                    # where data is a list of data series. count is used to index into data to select the correct data series for each subplot.
        axs[row,col].plot(x,data[count])
        count += 1         # This increments count so that the next subplot will use the next data series in data.

plt.show()
```
